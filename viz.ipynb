```python
plt.plot(losses)
```

```python
CONST = 64

def smooth(li):
    new_li = []
    for ix in range(len(li)):
        segment = li[max(0, ix-CONST):min(len(li), ix+CONST)]
        new_li.append(sum(segment)/len(segment))
    return new_li

smoothed = smooth(losses)

plt.plot(smoothed)

# ax = plt.subplot()

# ax.plot(smoothed[:15000], alpha=0.8, label="Without RLM")
# ax.plot(smooth(other)[:15000], alpha=0.8, label="With RLM")
# ax.legend()
# ax.set_title("Row-Level Masking Efficacy of Training MLM Loss")
# ax.set_xlabel("Batch #")
# ax.set_ylabel("MLM Loss")
# plt.show()
```
